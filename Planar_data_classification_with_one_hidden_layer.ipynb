1. Что такое веса и биасы в нейронных сетях?

Веса и смещения (биасы) — это параметры нейронной сети, которые определяют её выход.

Веса: Веса представляют собой силу связи между нейронами в разных слоях. Они умножаются на входные сигналы из предыдущего слоя. Большие веса указывают на более сильное влияние.

Смещения (биасы): Смещения добавляются к взвешенной сумме входов перед применением функции активации. Они позволяют нейрону активироваться даже если все взвешенные входы равны нулю, добавляя гибкости сети. Можно представить их как сдвиг кривой функции активации.

2. Как выглядят функции и как выглядят производные для функций активации нейронов sigmoid, tanh, ReLU?

Это распространенные функции активации:

Сигмоида: σ(x) = 1 / (1 + exp(-x)) Её производная: σ’(x) = σ(x)(1 - σ(x))

Гиперболический тангенс (tanh): tanh(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x)) Её производная: tanh’(x) = 1 - tanh²(x)

ReLU (линейная функция с исправлением): ReLU(x) = max(0, x) Её производная равна 1 при x > 0 и 0 при x ≤ 0 (она недифференцируема при x=0, но на практике используются субградиенты).

3. Что такое инициализация весов нейронной сети? Как именно она задаётся в лабораторной работе?

Инициализация весов — это процесс присвоения начальных значений весам нейронной сети перед началом обучения. Плохая инициализация может привести к медленному или не сходящемуся обучению. Распространенные методы включают случайную инициализацию (например, из нормального или равномерного распределения), инициализацию Xavier/Glorot и инициализацию He. Конкретный метод зависит от деталей вашего лабораторного задания.

4. Выпишите формулы для основных функций потерь (loss): среднеквадратичной ошибки (MSE), бинарной кроссэнтропии (BCE) и кроссэнтропии (CE). В каком случае какую функцию лучше всего использовать?

Среднеквадратичная ошибка (MSE): MSE = (1/n) * Σ(yᵢ - ŷᵢ)² где yᵢ — истинное значение, ŷᵢ — предсказанное значение, а n — количество выборок. Хорошо подходит для задач регрессии.

Бинарная кросс-энтропия (BCE): BCE = -(yᵢ * log(ŷᵢ) + (1 - yᵢ) * log(1 - ŷᵢ)) Используется для задач бинарной классификации (где выход — вероятность от 0 до 1).

Кросс-энтропия (CE): CE = -Σ yᵢ * log(ŷᵢ) Используется для задач многоклассовой классификации (где выход — это распределение вероятностей по нескольким классам). Обратите внимание, что часто она реализуется с функцией softmax перед вычислением потерь, чтобы гарантировать, что выходы являются корректным распределением вероятностей.

5. Как происходит процесс вычисления прямого прохода (forward propagation) и обратного распространения ошибки (error backpropagation) для двуслойной нейросети в матрично-векторном виде? Для чего они нужны?

Это включает матричные умножения и поэлементные операции. Подробное объяснение требует уравнений и матриц, которые сложно здесь воспроизвести. Однако основная идея:

Прямой проход: Входные данные распространяются через слои сети; веса умножаются на входы; добавляются смещения; применяются функции активации. Это генерирует предсказания.

Обратное распространение ошибки: Вычисляется разница между предсказанием и истинным значением (ошибка). Эта ошибка распространяется обратно через сеть для вычисления градиентов функции потерь относительно весов и смещений. Эти градиенты используются для обновления весов и смещений с помощью градиентного спуска, чтобы минимизировать ошибку в будущих предсказаниях. И то, и другое фундаментально для обучения нейронных сетей.

6. Что такое broadcasting матрично-векторных операций в numpy? Что означает параметр keepdims=True в функциях np.sum, np.mean и т.п.?

Broadcasting: Механизм в NumPy, который позволяет выполнять операции между массивами разных форм при определенных условиях. Фактически он расширяет меньший массив до размеров большего массива перед операцией.

keepdims=True: В агрегирующих функциях NumPy (таких как np.sum и np.mean), keepdims=True сохраняет уменьшенное(ые) измерение(я) в выходном массиве. Это важно для поддержания правильной формы массива в последующих вычислениях (особенно в многомерных случаях), предотвращая несоответствие форм и обеспечивая правильный broadcasting в дальнейших операциях.

7. В чём заключается обучение нейронной сети?

Обучение нейронной сети заключается в корректировке весов и смещений для минимизации разницы между её предсказаниями и фактическими целевыми значениями в обучающих данных. Это достигается итеративно с использованием обратного распространения ошибки и алгоритмов оптимизации, таких как градиентный спуск (или его варианты, такие как Adam, RMSprop и т. д.).

8. Что означает параметр скорости обучения (learning rate)? За что он отвечает, стоит ли его менять и каким его выбирать?

Скорость обучения — это гиперпараметр, который определяет размер шага при обновлении весов и смещений во время обучения. Меньшая скорость обучения приводит к меньшим обновлениям, что может привести к более стабильной, но медленной сходимости; в то время как большая скорость обучения приводит к большим обновлениям, что приводит к более быстрой, но потенциально менее стабильной сходимости (она может перепрыгнуть оптимальные значения и не сойтись). Оптимальная скорость обучения часто требует экспериментов (например, с использованием планировщиков скорости обучения). Нет универсального идеального значения; оно зависит от конкретного набора данных и архитектуры сети.

**Математически**:

Для одного примера $x^{(i)}$:
$$z^{[1] (i)} =  W^{[1]} x^{(i)} + b^{[1]}\tag{1}$$
$$a^{[1] (i)} = \tanh(z^{[1] (i)})\tag{2}$$
$$z^{[2] (i)} = W^{[2]} a^{[1] (i)} + b^{[2]}\tag{3}$$
$$\hat{y}^{(i)} = a^{[2] (i)} = \sigma(z^{ [2] (i)})\tag{4}$$
$$y^{(i)}_{prediction} = \begin{cases} 1 & \mbox{if } a^{[2](i)} > 0.5 \\ 0 & \mbox{otherwise } \end{cases}\tag{5}$$

Учитывая прогнозы всех примеров, вы также можете вычислить стоимость $J$ следующим образом:
$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large\left(\small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right)  \large  \right) \small \tag{6}$$

````python
# GRADED FUNCTION: initialize_parameters

def initialize_parameters(n_x, n_h, n_y):
    """
    Argument:
    n_x -- size of the input layer
    n_h -- size of the hidden layer
    n_y -- size of the output layer

    Returns:
    params -- python dictionary containing your parameters:
                    W1 -- weight matrix of shape (n_h, n_x)
                    b1 -- bias vector of shape (n_h, 1)
                    W2 -- weight matrix of shape (n_y, n_h)
                    b2 -- bias vector of shape (n_y, 1)
    """

    np.random.seed(2) # we set up a seed so that your output matches ours although the initialization is random.

    W1 = np.random.randn(n_h, n_x) * 0.01  # Матрица весов для скрытого слоя
    b1 = np.zeros((n_h, 1))                # Вектор смещения для скрытого слоя
    W2 = np.random.randn(n_y, n_h) * 0.01  # Матрица весов для выходного слоя
    b2 = np.zeros((n_y, 1))

    assert (W1.shape == (n_h, n_x))
    assert (b1.shape == (n_h, 1))
    assert (W2.shape == (n_y, n_h))
    assert (b2.shape == (n_y, 1))

    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}

    return parameters
````

### 4.3 - Полный цикл ####

**Задача**: Реализуйте `forward_propagation()`.

**Инструкции**:
- Посмотрите выше на математическое представление вашего классификатора.
- Вы можете использовать функцию `sigmoid()`. Он встроен (импортирован) в ноутбук.
- Вы можете использовать функцию `np.tanh()`. Это часть библиотеки numpy.
- Шаги, которые вам необходимо реализовать:
     1. Получите каждый параметр из словаря "parameters" (который является результатом `initialize_parameters()`), используя `parameters[..]`.
     2. Внедрить прямое распространение. Вычислите $Z^{[1]}, A^{[1]}, Z^{[2]}$ и $A^{[2]}$ (вектор всех ваших прогнозов на всех примерах в обучении). набор).
- Значения, необходимые для обратного распространения ошибки, хранятся в «кеше». «Кэш» будет передан в качестве входных данных для функции обратного распространения ошибки.

````python
# GRADED FUNCTION: forward_propagation

def forward_propagation(X, parameters):
    """
    Argument:
    X -- input data of size (n_x, m)
    parameters -- python dictionary containing your parameters (output of initialization function)

    Returns:
    A2 -- The sigmoid output of the second activation
    cache -- a dictionary containing "Z1", "A1", "Z2" and "A2"
    """
    # Retrieve each parameter from the dictionary "parameters"
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]

    # Реализуем прямое распространение для вычисления A2 (вероятности)
    Z1 = np.dot(W1, X) + b1         # Линейная часть для скрытого слоя
    A1 = np.tanh(Z1)               # Активация скрытого слоя
    Z2 = np.dot(W2, A1) + b2       # Линейная часть для выходного слоя
    A2 = 1 / (1 + np.exp(-Z2))

    assert(A2.shape == (1, X.shape[1]))

    cache = {"Z1": Z1,
             "A1": A1,
             "Z2": Z2,
             "A2": A2}

    return A2, cache
````

Теперь, когда вы вычислили $A^{[2]}$ (в переменной Python «`A2`»), которая содержит $a^{[2](i)}$ для каждого примера, вы можете вычислить функцию потерь (лосс) следующим образом:

$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large{(} \small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right) \large{)} \small\tag{13}$$

**Упражнение**. Реализуйте `compute_cost()` для вычисления значения стоимости $J$.

**Инструкция**:  
Есть много способов реализовать потерю перекрестной энтропии. Чтобы помочь вам, мы рассказываем, как бы мы реализовали
$- \sum\limits_{i=0}^{m}  y^{(i)}\log(a^{[2](i)})$:
```python
logprobs = np.multiply(np.log(A2),Y)
cost = - np.sum(logprobs)                # no need to use a for loop!
```

(вы можете использовать либо `np.multiply()`, а затем `np.sum()`, либо напрямую `np.dot()`).

````python
# GRADED FUNCTION: compute_cost

def compute_cost(A2, Y, parameters):
    """
    Computes the cross-entropy cost given in equation (13)

    Arguments:
    A2 -- The sigmoid output of the second activation, of shape (1, number of examples)
    Y -- "true" labels vector of shape (1, number of examples)
    parameters -- python dictionary containing your parameters W1, b1, W2 and b2

    Returns:
    cost -- cross-entropy cost given equation (13)
    """

    m = Y.shape[1] # number of example

    # Compute the cross-entropy cost
    logprobs = np.multiply(np.log(A2), Y) + np.multiply(np.log(1 - A2), 1 - Y)
    cost = - np.sum(logprobs) / m

    cost = np.squeeze(cost)     # makes sure cost is the dimension we expect.
                                # E.g., turns [[17]] into 17
    assert(isinstance(cost, float))

    return cost
````

- Совет:  
Чтобы вычислить dZ1, вам нужно вычислить $g^{[1]'}(Z^{[1]})$. Поскольку $g^{[1]}(.)$ — функция активации `tanh`, $a = g^{[1]}(z)$, то $g^{[1]'}(z) = 1- а^2$. Таким образом, вы можете вычислить $g^{[1]'}(Z^{[1]})$ используя `(1 - np.power(A1, 2))`.

````python
# GRADED FUNCTION: backward_propagation

def backward_propagation(parameters, cache, X, Y):
    """
    Implement the backward propagation using the instructions above.

    Arguments:
    parameters -- python dictionary containing our parameters
    cache -- a dictionary containing "Z1", "A1", "Z2" and "A2".
    X -- input data of shape (2, number of examples)
    Y -- "true" labels vector of shape (1, number of examples)

    Returns:
    grads -- python dictionary containing your gradients with respect to different parameters
    """
    m = X.shape[1]

    # Извлечение параметров W1 и W2 из словаря "parameters"
    W1 = parameters["W1"]
    W2 = parameters["W2"]

    # Извлечение A1 и A2 из словаря "cache"
    A1 = cache["A1"]
    A2 = cache["A2"]

    # Обратное распространение ошибки: вычисляем dW1, db1, dW2, db2
    dZ2 = A2 - Y                          # Ошибка на выходном слое
    dW2 = (1 / m) * np.dot(dZ2, A1.T)     # Градиент по W2
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)  # Градиент по b2

    dZ1 = np.dot(W2.T, dZ2) * (1 - np.power(A1, 2))  # Ошибка на скрытом слое
    dW1 = (1 / m) * np.dot(dZ1, X.T)                # Градиент по W1
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)  # Градиент по b1

    grads = {"dW1": dW1,
             "db1": db1,
             "dW2": dW2,
             "db2": db2}

    return grads
````

**Задача**: реализуйте правило обновления. Используйте градиентный спуск. Вам необходимо использовать (dW1, db1, dW2, db2) для обновления (W1, b1, W2, b2).

**Общее правило градиентного спуска**: $ \theta = \theta - \alpha \frac{\partial J }{ \partial \theta }$, где $\alpha$ — скорость обучения, а $\theta$ — параметр.

**Иллюстрация**: Алгоритм градиентного спуска с хорошей скоростью обучения (сходимость) и плохой скоростью обучения (расхождение). Изображения любезно предоставлены Адамом Харли.

````python
# GRADED FUNCTION: update_parameters

def update_parameters(parameters, grads, learning_rate = 1.2):
    """
    Updates parameters using the gradient descent update rule given above

    Arguments:
    parameters -- python dictionary containing your parameters
    grads -- python dictionary containing your gradients

    Returns:
    parameters -- python dictionary containing your updated parameters
    """
    # Retrieve each parameter from the dictionary "parameters"
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]

    # Извлечение градиентов из словаря "grads"
    dW1 = grads["dW1"]
    db1 = grads["db1"]
    dW2 = grads["dW2"]
    db2 = grads["db2"]

    # Обновление параметров
    W1 = W1 - learning_rate * dW1
    b1 = b1 - learning_rate * db1
    W2 = W2 - learning_rate * dW2
    b2 = b2 - learning_rate * db2

    parameters = {"W1": W1,
                  "b1": b1,
                  "W2": W2,
                  "b2": b2}

    return parameters
````

### 4.4 — Интегрируйте части 4.1, 4.2 и 4.3 в nn_model() ####

**Упражнение**: Создайте свою модель нейронной сети с помощью `nn_model()`.

**Инструкция**: Модель нейронной сети должна использовать предыдущие функции в правильном порядке.

````python
# GRADED FUNCTION: nn_model

def nn_model(X, Y, n_h, num_iterations = 10000, print_cost=False):
    """
    Arguments:
    X -- dataset of shape (2, number of examples)
    Y -- labels of shape (1, number of examples)
    n_h -- size of the hidden layer
    num_iterations -- Number of iterations in gradient descent loop
    print_cost -- if True, print the cost every 1000 iterations

    Returns:
    parameters -- parameters learnt by the model. They can then be used to predict.
    """

    np.random.seed(3)
    n_x = layer_sizes(X, Y)[0]
    n_y = layer_sizes(X, Y)[2]

    # Initialize parameters, then retrieve W1, b1, W2, b2. Inputs: "n_x, n_h, n_y". Outputs = "W1, b1, W2, b2, parameters".
    parameters = initialize_parameters(n_x, n_h, n_y)
    W1 = parameters["W1"]
    b1 = parameters["b1"]
    W2 = parameters["W2"]
    b2 = parameters["b2"]

    # Loop (gradient descent)
    for i in range(0, num_iterations):

        # Forward propagation. Inputs: "X, parameters". Outputs: "A2, cache".
        A2, cache = forward_propagation(X, parameters)

        # Cost function. Inputs: "A2, Y, parameters". Outputs: "cost".
        cost = compute_cost(A2, Y, parameters)

        # Backpropagation. Inputs: "parameters, cache, X, Y". Outputs: "grads".
        grads = backward_propagation(parameters, cache, X, Y)

        # Gradient descent parameter update. Inputs: "parameters, grads". Outputs: "parameters".
        parameters = update_parameters(parameters, grads)

        # Print the cost every 1000 iterations
        if print_cost and i % 1000 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))

    return parameters
````

### 4.5 Прогнозы

Используйте прямое распространение для прогнозирования результатов.

**Напоминание**: predictions = $y_{prediction} = \mathbb 1 \text{{activation > 0.5}} = \begin{cases}
      1 & \text{if}\ activation > 0.5 \\
      0 & \text{otherwise}
    \end{cases}$  
    
Например, если вы хотите установить элементы матрицы X значениями 0 и 1 на основе порогового значения, вы должны сделать:```X_new = (X > threshold)```

````python
# GRADED FUNCTION: predict

def predict(parameters, X):
    """
    Using the learned parameters, predicts a class for each example in X

    Arguments:
    parameters -- python dictionary containing your parameters
    X -- input data of size (n_x, m)

    Returns
    predictions -- vector of predictions of our model (red: 0 / blue: 1)
    """

    # Computes probabilities using forward propagation, and classifies to 0/1 using 0.5 as the threshold.
    A2, cache = forward_propagation(X, parameters)
    predictions = (A2 > 0.5).astype(int)

    return predictions
````
